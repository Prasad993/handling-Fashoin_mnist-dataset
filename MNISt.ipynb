```
import tensorflow as tf
import keras
```

```
#importing the fashoin MNIST data
fashion_mnist=tf.keras.datasets.fashion_mnist
```

```
# loading the data
(train_images,train_labels),(test_images,test_labels)=fashion_mnist.load_data()
```

```
#Scaling the data
train_images=train_images/255
test_images=test_images/255
```

```
train_images.shape
```

```
train_images=train_images.reshape(len(train_images),28,28,1)
test_images=test_images.reshape(len(test_images),28,28,1)
```

```
#model building starts
cnn=keras.models.Sequential()
```

```
#Convolution layer 1 with 32 filters of 3*3 and with relu activation fuv=ction with input_shape of (28,28,1) and weight_initalizer is glorot_uniform
cnn.add(keras.layers.Conv2D(filters=32,activation="relu",kernel_size=(3,3),kernel_initializer='glorot_uniform',input_shape=[28,28,1]))
```

```
# Pooling layer1 with pools_ize of 2*2 and stride of2
cnn.add(keras.layers.MaxPool2D(pool_size=(2,2),strides=2))
```

```
#Convolution layer2 with 32 filters of 3*3 and with relu activation function  and weight_initalizer is glorot_uniform

cnn.add(keras.layers.Conv2D(filters=32,activation="relu",kernel_size=(3,3),kernel_initializer='glorot_uniform'))
```

```
# Pooling layer1 with pools_ize of 2*2 and stride of2

cnn.add(keras.layers.MaxPool2D(pool_size=(2,2),strides=2))
```

```
#Flatten layer
cnn.add(keras.layers.Flatten())
```

```
#fully connected layer with 10 units and softmax activation fuction
cnn.add(keras.layers.Dense(units=10,activation="softmax"))
```

```
# compiling the model with adam optimizer and loss function as sparse_categorical_crossentropy
cnn.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```
#model summary
cnn.summary()
```

```
#model training got validation accuracy as 94.7%
cnn.fit(x=train_images,y=train_labels,epochs=25)
```

```

#evaluationg the model 
score=cnn.evaluate(test_images, test_labels, verbose=0)
```

```
print('test loss: {:.4f}'.format(score[0]))
print(' test acc: {:.4f}'.format(score[1]))
```
